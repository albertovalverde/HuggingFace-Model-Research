```python
# 1️⃣ Install dependencies silently
!git lfs install
!git clone https://huggingface.co/hexgrad/Kokoro-82M
%cd Kokoro-82M
!apt-get -qq -y install espeak-ng > /dev/null 2>&1
!pip install -q phonemizer torch transformers scipy munch

# 2️⃣ Build the model and load the default voicepack
from models import build_model
import torch
device = 'cuda' if torch.cuda.is_available() else 'cpu'
MODEL = build_model('kokoro-v0_19.pth', device)
VOICE_NAME = [
    'af', # Default voice is a 50-50 mix of Bella & Sarah
    'af_bella', 'af_sarah', 'am_adam', 'am_michael',
    'bf_emma', 'bf_isabella', 'bm_george', 'bm_lewis',
    'af_nicole', 'af_sky',
][0]
VOICEPACK = torch.load(f'voices/{VOICE_NAME}.pt', weights_only=True).to(device)
print(f'Loaded voice: {VOICE_NAME}')

# 3️⃣ Call generate, which returns 24khz audio and the phonemes used
from kokoro import generate
text = """
In the case of John Doe v. XYZ Corporation, the Court has reviewed all presented evidence, including the testimonies of the plaintiff and the defendant, along with the documentary evidence submitted during the proceedings.

After thorough deliberation, the Court finds that the defendant, XYZ Corporation, is liable for breach of contract under the terms outlined in the agreement signed on June 15, 2022. The plaintiff, John Doe, has proven by clear and convincing evidence that XYZ Corporation failed to deliver the goods as stipulated in the contract, causing substantial financial harm.

The Court hereby orders that XYZ Corporation shall pay the plaintiff, John Doe, the amount of $250,000 in damages, which includes both compensatory and consequential damages. Furthermore, the defendant is to cover the legal costs incurred by the plaintiff.

This judgment is final and binding, subject only to the right of appeal within 30 days from the date of this decision.
"""
audio, out_ps = generate(MODEL, text, VOICEPACK, lang=VOICE_NAME[0])

# Language is determined by the first letter of the VOICE_NAME:
# 🇺🇸 'a' => American English => en-us
# 🇬🇧 'b' => British English => en-gb

# 4️⃣ Display the 24khz audio and print the output phonemes
from IPython.display import display, Audio
display(Audio(data=audio, rate=24000, autoplay=True))
print(out_ps)
```
